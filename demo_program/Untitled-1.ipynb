```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

Image.MAX_IMAGE_PIXELS = 99999999999  

floating_image = Image.open("/Users/clark/Downloads/Sp13 13196 Fibrosis 60% Data/Array Nuclear Marker Test.tif")
target_image = Image.open("/Users/clark/Downloads/Sp13 13196 Fibrosis 60% Data/Sp13 13196 Fibrosis 60% Protein Panel.ome.tif")
target_image.seek(1)
```

```python
# Crop both images to the specified region
crop_region = (1000, 1000, 3000, 3000)  # (left, top, right, bottom)
floating_image = floating_image.crop(crop_region)
target_image = target_image.crop(crop_region)
```

```python
from microfilm.microplot import microshow

microshow(
    images=[np.array(floating_image), np.array(target_image)], fig_scaling=25,
    cmaps=['pure_red', 'pure_green'])
```
